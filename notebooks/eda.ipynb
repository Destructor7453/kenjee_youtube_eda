- Need to visualize the distributions of the metrics we're interested in for audience engagement

```python
metrics = [
    'Views',
    'Likes',
    'Av­er­age view dur­a­tion',
    'Av­er­age per­cent­age viewed (%)'
]
df.columns
```

```python
plt.style.use('seaborn-v0_8')
for col in metrics:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], bins=30, kde=True)
    plt.title(f'{col} Distribution')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
```

The distribution of 'Views' and 'Likes' is very skewed to the right (Right-skewed), since the tails point to the right. This means that there must be outliers in the data and we have to clean for them. We can also try to scale the distribution via a log scale (seen below). The distribution of 'Average view duration' is not normal, so we'll need to normalize it to continue with our analysis. 

```python
for col in ['Views', 'Likes']:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], bins=30, kde=True, log_scale=True)
    plt.title(f'{col} Distribution (Log Scale)')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
```

The purpose of log-scaling in a distribution is to compress the large scale range of data into a more manageable range, which makes it easier to visualize and analyze distributions. It also transforms multiplicative relationships into additive ones, which are useful in linearization of exponential growth. The features 'Views' and 'Likes' are log-normally distributed, the logarithms of 'Views' and 'Likes' follow a normal distribution. 

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
sns.boxplot(x=df['Views'])
plt.title('Views Boxplot')

plt.subplot(1,2,2)
sns.boxplot(x=df['Likes'])
plt.title('Likes Boxplot')

plt.tight_layout()
plt.show()
```

In the boxplots above, we can see that there are outliers in the 'Views' and 'Likes' features, which are common for video platforms like YouTube. This is because some videos may do average or as expected while others may do really well (i.e. viral) or just have an extraordinary engagement because of the topic. To deal with these outliers, I am going to cap the data versus removing them so I do not lose data. 

```python
def cap_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    data[column] = np.where(data[column] > upper_bound, upper_bound,
                            np.where(data[column] < lower_bound, lower_bound, data[column]))
    return data

df_capped = df.copy()
df_capped = cap_outliers_iqr(df_capped, 'Views')
df_capped = cap_outliers_iqr(df_capped, 'Likes')
```

```python
sns.histplot(df_capped['Views'], bins=30, kde=True)
plt.show()
```

Even dealing with the outliers, the data is still highly skewed (to the right), so we still need to log-scale the distribution to make it normal. 

```python
df_transformed = df.copy()
df_transformed['Views_log'] = np.log1p(df_transformed['Views'])
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df_transformed['Views_log'], bins=30, kde=True)
plt.title('Views (Log Transformed)')
plt.show()
```

Similarly for 'Likes':

```python
sns.histplot(df_capped['Likes'], bins=30, kde=True)
plt.show()
```

```python
df_transformed['Likes_log'] = np.log1p(df_transformed['Likes'])
plt.figure(figsize=(6,4))
sns.histplot(df_transformed['Likes_log'], bins=30, kde=True)
plt.title('Likes (Log Transformed)')
plt.show()
```

There is slight difference in the distributions when we cleaned/treated the dta set for outliers. 

Below is a side-by-side comparision of the before and after:

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

sns.histplot(df['Views'], bins=30, kde=True, ax=axes[0,0])
axes[0,0].set_title('Views (Raw)')

sns.histplot(df_transformed['Views_log'], bins=30, kde=True, ax=axes[0,1])
axes[0,1].set_title('Views (Log Transformed)')

sns.histplot(df['Likes'], bins=30, kde=True, ax=axes[1,0])
axes[1,0].set_title('Likes (Raw)')

sns.histplot(df_transformed['Likes_log'], bins=30, kde=True, ax=axes[1,1])
axes[1,1].set_title('Likes (Log Transformed)')

plt.tight_layout()
plt.show()
```

Now let's look at the features as a whole, namely look at if there are any relations between the different features in the dataset:

```python
numeric_df = df_transformed.select_dtypes(include=['float64', 'int64'])
numeric_df.head()
corr_matrix = numeric_df.corr(method='pearson')
corr_matrix
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(
    corr_matrix,
    annot=True,
    fmt=".2f",
    cmap='coolwarm',
    linewidths=0.5,
    cbar_kws={'shrink': 0.8}
)
plt.title('Pearson Correlation Matrix')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.regplot(
    data=df_transformed,
    x='Views_log',
    y='Likes_log',
    scatter_kws={'alpha':0.6},
    line_kws={'color':'red'}
)
plt.title('Views vs Likes (with Regression Line)')
plt.xlabel('Log(Views)')
plt.ylabel('Log(Likes)')
plt.tight_layout()
plt.show()
```
